```python
import glob
```

```python
base_path = '../../Benchmarks/SCDISENTANGLE'
```

# Kang Counterfactuals

```python
criterion_files = glob.glob(
    f'{base_path}/Kang/weights/kang/*/*/r2_mean_criterion_10_all/value.txt'
)
len(criterion_files)
```

```python
all_values = []
sizes = []
for file in criterion_files:
    with open(file, 'r') as f:
        values = f.readlines()
    values = [float(x.replace('\n', '')) for x in values]
    all_values.extend(values)
    sizes.append(len(values))
```

```python
len(all_values), len(sizes), min(all_values), max(all_values)
```

```python
for _val in all_values:
    assert _val < 1 and _val > 0
```
