```python
import scanpy as sc
import numpy as np
import os
import glob
from tqdm import tqdm
import gc
```

```python
def check_nan(data_name):
    stats = {'min':[], 'max': []}
    
    pred_files = glob.glob(f'../../Benchmarks/BIOLORD/{data_name}/predictions/*.h5ad')
    print('Number of predictions: ,', len(pred_files))
    for file in tqdm(pred_files):
        adata = sc.read_h5ad(file)
        n_nan = np.isnan(adata.X).sum()
        assert n_nan == 0
        assert np.isfinite(adata.X).all()
        assert np.min(adata.X) >= -1e6  # optional guardrail

        stats['min'].append(adata.X.min())
        stats['max'].append(adata.X.max())
        
        del adata
        gc.collect()
        
    print('Test passed for: ', data_name)
    return stats
```

```python
kang_stats = check_nan(data_name='Kang')
```

```python
liver_stats = check_nan(data_name='Liver')
```

```python
np.min(liver_stats['min']), np.mean(liver_stats['min'])
```
