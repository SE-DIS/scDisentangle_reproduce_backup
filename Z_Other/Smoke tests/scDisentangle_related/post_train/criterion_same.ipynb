```python
import os
import glob
```

```python
base_path = '../../../../Benchmarks/SCDISENTANGLE'
```

```python
def is_same(data_name):
    old_weights = glob.glob(
        f'{base_path}/{data_name}/Jan 14/weights/**/epoch_nb.txt',
        recursive=True
    )
    print('Evaluating', len(old_weights), 'file')
    for file in old_weights:
        with open(file, 'r') as f:
            old_epochs=f.readlines()
            old_epochs=[x.replace('\n', '') for x in old_epochs]
            old_last_epoch = old_epochs[-1]

        current_file = file.replace('Jan 14/', '')
        with open(current_file, 'r') as f:
            current_epochs=f.readlines()
            current_epochs=[x.replace('\n', '') for x in current_epochs]
            current_last_epoch = current_epochs[-1]

        assert current_last_epoch == old_last_epoch

    print('Assertions passed for', data_name)
```

```python
is_same('Kang')
```

```python
is_same('Liver')
```

```python
is_same('Prostate')
```

```python
is_same('Norman')
```
