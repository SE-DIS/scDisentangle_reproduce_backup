```python
import numpy as np
import torch
import random
from icecream import ic
```

```python
def set_seed(seed):
    ic('Setting seed to', seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
```

# Read data

```python
data_path = '/data/Experiments/Benchmark/SCDISENTANGLE_REPRODUCE/Datasets/preprocessed_datasets/norman.h5ad'
```

```python
adata = sc.read(data_path)

adata.X = adata.layers['counts'].copy()
adata.X = adata.X.toarray()

adata
```

# Make dose_value and rank_genes_groups_cov and cov_cond

```python
# Dose value
pert_labels = adata.obs['cond_harm'].tolist()
dose_values = []

for _l in pert_labels:
    if '+' in _l:
        dose_values.append('1.0+1.0')
    else:
        dose_values.append('1.0')

adata.obs['dose_value'] = dose_values
adata.obs['dose_value'] = adata.obs['dose_value'].astype('category')
```

```python
adata.obs['dose_value'].value_counts()
```

```python
# cov_cond

cov_conds = ['K562_' + x for x in pert_labels]
adata.obs['cov_cond'] = cov_conds
```

```python
adata.obs['cov_cond']
```

```python
# rank_genes_groups_cov
```

```python
adata.uns['rank_genes_groups_cov'] = {'K562_' +k:v for k,v in adata.uns['rank_genes_groups'].items()}
```

```python
adata.uns['rank_genes_groups_cov'].keys()
```

```python
adata.uns['rank_genes_groups_cov']['K562_AHR+FEV']
```

# Setup Anndata

```python
cpa.CPA.setup_anndata(adata, 
                      perturbation_key='cond_harm',
                      control_group='ctrl',
                      dosage_key='dose_value',
                      categorical_covariate_keys=['cell_type'],
                      is_count_data=True,
                      deg_uns_key='rank_genes_groups_cov',
                      deg_uns_cat_key='cov_cond',
                      max_comb_len=2,
                     )
```

```python
adata.obs['cond_harm']
```

# Parameters

```python
model_params = {
    "n_latent": 32,
    "recon_loss": "nb",
    "doser_type": "linear",
    "n_hidden_encoder": 256,
    "n_layers_encoder": 4,
    "n_hidden_decoder": 256,
    "n_layers_decoder": 2,
    "use_batch_norm_encoder": True,
    "use_layer_norm_encoder": False,
    "use_batch_norm_decoder": False,
    "use_layer_norm_decoder": False,
    "dropout_rate_encoder": 0.2,
    "dropout_rate_decoder": 0.0,
    "variational": False,
    "seed": 8206,
}

trainer_params = {
    "n_epochs_kl_warmup": None,
    "n_epochs_adv_warmup": 50,
    "n_epochs_mixup_warmup": 10,
    "n_epochs_pretrain_ae": 10,
    "mixup_alpha": 0.1,
    "lr": 0.0001,
    "wd": 3.2170178270865573e-06,
    "adv_steps": 3,
    "reg_adv": 10.0,
    "pen_adv": 20.0,
    "adv_lr": 0.0001,
    "adv_wd": 7.051355554517135e-06,
    "n_layers_adv": 2,
    "n_hidden_adv": 128,
    "use_batch_norm_adv": True,
    "use_layer_norm_adv": False,
    "dropout_rate_adv": 0.3,
    "step_size_lr": 25,
    "do_clip_grad": False,
    "adv_loss": "cce",
    "gradient_clip_value": 5.0,
}
```

```python
# Make the split train valid ood
```

```python
_split_dict.keys()
```

```python
train_mask = adata.obs['cond_harm'].isin(_split_dict['1']['train'])
val_mask = adata.obs['cond_harm'].isin(_split_dict['1']['val'])
ood_mask = adata.obs['cond_harm'].isin(_split_dict['1']['ood'])

adata.obs['split'] = None
adata.obs['split'][train_mask] = 'train'
adata.obs['split'][val_mask] = 'valid'
adata.obs['split'][ood_mask] = 'ood'
```

```python
adata.obs['split'].value_counts()
```

```python
adata[adata.obs['split'] == 'ood'].obs['cond_harm'].value_counts()
```

# Init model

```python
model = cpa.CPA(adata=adata, 
                split_key='split',
                train_split='train',
                valid_split='valid',
                test_split='ood',
                **model_params,
               )
```

# Train

```python
model.train(max_epochs=2000,
            use_gpu=True, 
            batch_size=2048,
            plan_kwargs=trainer_params,
            early_stopping_patience=5,
            check_val_every_n_epoch=5,
            save_path='weights/',
           )
```

```python
cpa.pl.plot_history(model)
```

```python
pert_name = 'CEBPB+PTPN12' #'CEBPB+MAPK1'
```

```python
degs = adata.uns['rank_genes_groups'][pert_name]
degs_indices = [adata.var_names.get_loc(x) for x in degs]
```

```python
adata_subset = adata[adata.obs['cond_harm'] == pert_name]
ctrl = adata[adata.obs['cond_harm'] == 'ctrl'].layers['X_true']

print(adata_subset.shape, ctrl.shape)

for n_degs in [200, 100, 50, 10]:
    degs_mask = degs_indices[:n_degs]

    pred = adata_subset.layers['CPA_pred']
    gt = adata_subset.layers['X_true']
    
    r2 = r2_score(
        gt.mean(axis=0)[degs_mask],
        pred.mean(axis=0)[degs_mask]
    )

    pear, _ = pearsonr(
        gt.mean(axis=0)[degs_mask],
        pred.mean(axis=0)[degs_mask]
    )

    pear_delta, _ = pearsonr(
        gt.mean(axis=0)[degs_mask] - ctrl.mean(axis=0)[degs_mask],
        pred.mean(axis=0)[degs_mask] - ctrl.mean(axis=0)[degs_mask]
    )

    r2_delta = r2_score(
        gt.mean(axis=0)[degs_mask] - ctrl.mean(axis=0)[degs_mask],
        pred.mean(axis=0)[degs_mask] - ctrl.mean(axis=0)[degs_mask]
    )

    print(n_degs, 'R2 score: ', r2, 'Pearson : ', pear, 'Pear Delta: ', pear_delta, 'R2_delta: ', r2_delta)
```

```python
adata.X = adata.layers['counts'].copy()
adata_subset = adata[adata.obs['cond_harm'] == 'ctrl']#[subset_indices,:]

try:
    del adata.obsm['CPA_pred']
except:
    pass

try:
    del adata.layers['CPA_pred']
except:
    pass
    
var_to_predict = pert_name
adata_pred = adata_subset.copy()

# Make cond_harm_org
adata_pred.obs['cond_harm_org'] = adata_pred.obs['cond_harm'].copy()

# Set cond_harm and cond_harm_pred to var_to_predict
adata_pred.obs['cond_harm_pred'] = [var_to_predict] * adata_pred.shape[0]
adata_pred.obs['cond_harm'] = [var_to_predict] * adata_pred.shape[0]

# Set cov_cond to cell_type + _ + var_to_predict
cov_cond_pred = []
for i in range(adata_pred.shape[0]):
    _name = adata_pred.obs['cell_type'][i] + '_' + adata_pred.obs['cond_harm'][i]
    cov_cond_pred.append(_name)

adata_pred.obs['cov_cond'] = cov_cond_pred

        # Set dose value
adata_pred.obs['dose_value'] = adata_pred.obs['cond_harm'].apply(lambda x: '+'.join(['1.0' for _ in x.split('+')]))

        # Setup anndata
cpa.CPA.setup_anndata(
            adata_pred, 
            perturbation_key='cond_harm',
            control_group='ctrl',
            dosage_key='dose_value',
            categorical_covariate_keys=['cell_type'],
            is_count_data=True,
            deg_uns_key='rank_genes_groups_cov',
            deg_uns_cat_key='cov_cond',
            max_comb_len=2,
                )

        # Predict
model.predict(adata_pred, batch_size=2048)
```
