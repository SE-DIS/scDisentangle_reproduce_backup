# Export results

```python
import os
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import hygeia.utils.hplots as hp
import metrics_tools as mt
```

```python
data_name = 'Norman'
cond_key = 'condition'
control_name = 'ctrl'
custom_name = 'combinatorially_seen'
ood_labels = double_perts

all_results = []
for ood_l in ood_labels:
    all_results.append(f'{ood_l}')
```

```python
base_path = f'results/{data_name}/{custom_name}'
#save_path = f'/data/scDisentangle figures/{data_name}/Benchmarks'
#save_path = f'/data/baselines/{data_name}/{custom_name}'
#os.makedirs(save_path, exist_ok=True)
```

```python
save_path = f'/data/baselines/{data_name}/{custom_name}' #f'/data/scDisentangle figures/{data_name}/Benchmarks'
os.makedirs(save_path, exist_ok=True)
```

```python
colors = {
    'scDisentangle': '#d62728',
    'Match Mean': '#698396',
    'GEARS': '#00FFFF',
    'CPA': '#CC79A7',
    #'Pert Mean': '#00FFFF', #'#698396',
    
    }
```

```python
folder_paths = {
     'scDisentangle': f'{base_path}/SCDISENTANGLE/',
     'GEARS': f'{base_path}/GEARS',
     'CPA': f'{base_path}/CPA/', 
     'Match Mean': f'{base_path}/Match Mean',
     #'Pert Mean': f'{base_path}/Pert Mean',
}
```

```python
n_degs_list = ['200', '100', '50', '20', '10']
metric_name = 'Wasserstein'

metrics = mt.get_results(
    folder_paths, 
    metric_name=metric_name, 
    perts = all_results, 
    n_degs_list=n_degs_list
)

pvals, stars = mt.compute_pvalues(
    metrics_dict=metrics,
    metric_name=metric_name,
    higher_is_better=None,
    anchor_method='scDisentangle',
    best_method_only=True
)

records = hp.plot_data(
            metrics,
            plot_type='barplot', 
            save_path=None, #f'{save_path}/{metric_name}',
            y_name=metric_name, 
            legend=True, 
            title=metric_name,
            colors=colors,
            show_grid=False,
            sort_order=None,
            stars=stars,
            figsize=(4, 4)
        )
```
