# Export results

```python
import os
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import hygeia.utils.hplots as hp
import metrics_tools as mt
```

```python
data_name = 'Tabula'
cov_key = 'cell_type'
cond_key = 'specie'
stim_name = 'sapiens'
control_name = 'muris'
covs = ['pulmonary alveolar type 2 cell']

all_results = []
for cov in covs:
    for i in range(1, 11):
        all_results.append(f'{cov}_{i}')
```

```python
base_path = f'results/{data_name}'
```

```python
colors = {
    'scDisentangle': '#d62728',
    #'Perturbed Baseline': '#698396',
    #'Control Baseline': '#00FFFF',
    'CPA': '#CC79A7',
    'scGEN': '#E69F00',
    'ScDisInFact': '#0072B2',
    'Biolord': '#009E73',
    }
```

```python
folder_paths = {
     'scDisentangle': f'{base_path}/SCDISENTANGLE/',
     #'Perturbed Baseline': f'{base_path}/Pert Mean',
     #'Control Baseline': f'{base_path}/Ctrl Mean',
     'ScDisInFact': f'{base_path}/SCDISINFACT/',
     'CPA': f'{base_path}/CPA/',
     'scGEN': f'{base_path}/SCGEN/',
     'Biolord': f'{base_path}/BIOLORD/', 
}
```

```python
save_path = f'/data/scDisentangle figures/{data_name}/Benchmarks'
os.makedirs(save_path, exist_ok=True)
```

```python
n_degs_list = ['200', '100', '50', '20', '10']
metric_name = 'Single-cell Identity preservation'

metrics = mt.get_results(
    folder_paths, 
    metric_name=metric_name, 
    perts = all_results, 
    n_degs_list=n_degs_list
)

pvals, stars = mt.compute_pvalues(
    metrics,
    metric_name=metric_name,
    higher_is_better=True,
    anchor_method='scDisentangle'
)

records = hp.plot_data(
            metrics,
            plot_type='barplot', 
            save_path=None, #f'{save_path}/LEGEND',
            y_name=metric_name, 
            legend=True, 
            title=metric_name,
            colors=colors,
            show_grid=False,
            sort_order=None,
            stars=stars
        )
```

```python
records = hp.plot_data(
            metrics,
            plot_type='barplot', 
            save_path=None,
            y_name=metric_name, 
            legend=True, 
            title=metric_name,
            colors=colors,
            show_grid=False,
            sort_order=None,
            stars=stars
        )
```
