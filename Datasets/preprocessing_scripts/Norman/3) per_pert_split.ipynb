```python
import json
```

# Scenario 1

```python
per_pert_splits = {}
with open('../../preprocessed_datasets/norman_splits_1.json', 'r') as f:
    j_dict_1 = json.load(f)
    
split_indices = sorted(j_dict_1.keys(), key=int)  # optional

for split_idx in split_indices:
    split_dict = j_dict_1[split_idx]
    train_perts = [p for p in split_dict['train'] if p != 'ctrl']  # copy

    for ood in split_dict['ood']:
        per_pert_splits[ood] = train_perts.copy()

with open('../../preprocessed_datasets/per_pert_splits_1.json', 'w') as f:
    json.dump(per_pert_splits, f)
```
