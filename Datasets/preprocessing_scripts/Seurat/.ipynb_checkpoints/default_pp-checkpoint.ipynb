```python
immature_cells = [
    "CD4 Naive",
    "CD8 Naive",
    "CD8 Naive_2",
    "CD4 TCM_1",
    "CD4 TCM_2",
    "CD4 TCM_3",
    "CD8 TCM_1",
    "CD8 TCM_2",
    "CD8 TCM_3",
    "B naive kappa",
    "B naive lambda",
    "B intermediate kappa",
    "B intermediate lambda",
    "NK_CD56bright",
    "NK Proliferating",
    "CD4 Proliferating",
    "CD8 Proliferating",
    "ASDC_mDC",
    "ASDC_pDC"
]
```

##### Link: Seurat

# Read and preprocess data

```python
data_path = '../../original_datasets/Seurat/seurat_pbmc.h5ad'
```

```python
adata = sc.read_h5ad(data_path)

adata.obs['celltype.l1'] = adata.obs['celltype.l1'].astype(str)
adata.obs['celltype.l1'][adata.obs['celltype.l1'] == 'Mono'] = adata[adata.obs['celltype.l1'] == 'Mono'].obs['celltype.l2'].copy()
adata.obs['celltype.l1'][adata.obs['celltype.l1'] == 'other'] = adata[adata.obs['celltype.l1'] == 'other'].obs['celltype.l2'].copy()
adata = adata[~adata.obs['celltype.l1'].isin(['HSPC', 'Platelet', 'Eryth'])]
adata = adata[~adata.obs['celltype.l3'].isin(immature_cells)]

print(adata.X.max(), adata.X.min())

adata.layers['counts'] = adata.X.copy()
adata.raw = adata.copy()
adata.uns = {}

sc.pp.filter_cells(adata, min_counts=100)
sc.pp.filter_genes(adata, min_counts=5)

sc.pp.normalize_total(
    adata, 
    target_sum=1e4, 
    exclude_highly_expressed=True
    )
sc.pp.log1p(adata)

adata.shape, adata.X.max(), adata.X.min()
```

```python
sc.pp.highly_variable_genes(
    adata, 
    n_top_genes=5000, 
    subset=True
)
```

# Create cell ids

```python
adata.obs['sc_cell_ids'] = list(range(adata.shape[0]))
```

# Write

```python
adata.X = adata.layers['counts'].copy()
```

```python
adata.write_h5ad('../../preprocessed_datasets/seurat.h5ad')
```
