# Imports

```python
import os
import torch
import scanpy as sc
import numpy as np
from scdisentangle.train.tools import get_trainer, set_seed
```

# Set seed

```python
seed_nb = 42
set_seed(seed_nb)
```

# Get Trainer

```python
yaml_path = 'configs/leukemia_disentangle.yaml'
```

```python
trainer = get_trainer(yaml_path, wandb_log=True, seed_nb=seed_nb)
```

# Train

```python
trainer.train()
```

# Load best weights

```python
weights_path = 'weights/MIG_BINNED_dis_latent_stack_Sample_id_train'
trainer.load_weights(weights_path)
```

# Get Latent

```python
counterfactual_dict = {'Sample_id': 'patient1_IP'}
```

# Inference

```python
adata = trainer.predict(
    trainer.dataset.data.copy(), 
    counterfactual_dict=counterfactual_dict, 
    bs=256
)
```

```python
adata_latent = sc.AnnData(
    adata.obsm['dis_latent_stack'].copy(),
    obs=adata.obs.copy()
)
```

```python
sc.pp.neighbors(adata_latent)
```

```python
sc.tl.umap(adata_latent)
```

```python
sc.pl.umap(
    adata_latent,
    color=['Sample_id_org'],
)
```

```python
sc.pl.umap(
    adata_latent,
    color=['cell_type'],
)
```
