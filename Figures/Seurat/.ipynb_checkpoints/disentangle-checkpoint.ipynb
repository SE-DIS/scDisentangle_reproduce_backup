# Imports

```python
import os
import torch
import scanpy as sc
import numpy as np
from scdisentangle.train.tools import get_trainer, set_seed
```

# Set seed

```python
seed_nb = 42
set_seed(seed_nb)
```

# Get Trainer

```python
yaml_path = 'configs/seurat_disentangle.yaml'
```

```python
trainer = get_trainer(yaml_path, wandb_log=True, seed_nb=seed_nb)
```

# Train

```python
trainer.train()
```

# Load best weights

```python
weights_path = 'weights/MIG_BINNED_dis_latent_stack_donor_train'
trainer.load_weights(weights_path)
```

# Get Latent

```python
counterfactual_dict = {'donor': 'P7'}
```

# Inference

```python
adata = trainer.predict(
    trainer.dataset.data.copy(), 
    counterfactual_dict=counterfactual_dict, 
    bs=256
)
```
