```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_method_scores(scores, colors, metric_label="Correlation", savepath=None, dpi=300):
    """
    scores: dict, e.g. corr or mi
    colors: dict mapping method -> color hex
    metric_label: label for the y-axis (e.g. 'Correlation' or 'Mutual information')
    savepath: if not None, path to save the figure
    dpi: resolution for saving
    """
    methods = list(scores.keys())
    values = np.array([scores[m] for m in methods])
    bar_colors = [colors[m] for m in methods]

    plt.style.use("seaborn-v0_8-white")

    fig, ax = plt.subplots(figsize=(3.0, 3.0))  # small but high-res, journal friendly

    # Bars
    bar_width = 0.7
    bars = ax.bar(
        x=np.arange(len(methods)),
        height=values,
        color=bar_colors,
        edgecolor="black",
        linewidth=1.0,
        width=bar_width,
    )

    # Aesthetics: labels and ticks
    ax.set_xticks(np.arange(len(methods)))
    ax.set_xticklabels(methods, rotation=30, ha="right",
                       fontsize=10, fontweight="bold")
    ax.set_ylabel(metric_label, fontsize=11, fontweight="bold")
    ax.set_xlabel("")  # cleaner look

    # Tight y-limits with a little headroom
    ymin = 0
    ymax = max(values) * 1.15 if max(values) > 0 else 1
    ax.set_ylim(ymin, ymax)

    # Remove gridlines
    ax.yaxis.grid(False)
    ax.xaxis.grid(False)

    # Clean spines
    for spine in ["top", "right"]:
        ax.spines[spine].set_visible(False)
    ax.spines["left"].set_linewidth(0.9)
    ax.spines["bottom"].set_linewidth(0.9)

    # Tick params
    ax.tick_params(axis="both", which="both",
                   labelsize=9, width=0.8, length=4)

    # Annotate bars with values (slightly larger, bold)
    for bar, val in zip(bars, values):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            val + (ymax * 0.01),
            f"{val:.3f}",
            ha="center",
            va="bottom",
            fontsize=8,
            fontweight="bold",
        )

    fig.tight_layout(pad=0.5)

    if savepath is not None:
        fig.savefig(savepath, dpi=dpi, bbox_inches="tight")
    return fig, ax
```

```python
with open('average_transition_castration.txt', 'r') as f:
    cast_values = f.readlines()
    cast_values = [x.replace('\n', '') for x in cast_values][:50]
    cast_values = [x for x in cast_values if not x in ['']]
    cast_values = [float(x) for x in cast_values]
    cast_mean = np.mean(cast_values)

with open('average_transition_regeneration.txt', 'r') as f:
    reg_values = f.readlines()
    reg_values = [x.replace('\n', '') for x in reg_values]
    reg_values = [x for x in reg_values if not x in ['']][:50]
    reg_values = [float(x) for x in reg_values]
    reg_mean = np.mean(reg_values)
```

```python
np.mean(reg_values)
```

```python
from scipy.stats import wilcoxon
wilcoxon(reg_values, cast_values)
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(reg_values, cast_values)
```

```python
from scipy.stats import mannwhitneyu

stat, p_value = mannwhitneyu(
    reg_values,
    cast_values,
    alternative='greater'  # cast > reg
)

print("U statistic:", stat)
print("p-value:", p_value)
```

```python
avg_prob_transition = {
    'Regeneration': reg_mean, #0.017517653994449715,
    'Castration': cast_mean #0.008427041744209156
}

colors = {
    'Regeneration': '#d62728',
    'Castration': '#00FFFF',
    }
```

```python
plot_method_scores(
    avg_prob_transition, 
    colors, 
    metric_label="Average Transition Probability",
    savepath=f"{save_path}avg_transition_prob.png", 
    dpi=600
)
```
