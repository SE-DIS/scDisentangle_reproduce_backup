```python
import pandas as pd
import numpy as np
from scipy.sparse import issparse

def get_cluster_transition_df(adata, tm):
    # Get cluster labels and unique clusters
    clusters = adata.obs['clusters'].values
    unique_clusters = np.unique(clusters)
    n_clusters = len(unique_clusters)
    
    # Map each cluster to its cell indices
    cluster_to_indices = {cluster: np.where(clusters == cluster)[0] for cluster in unique_clusters}
    
    # Initialize cluster transition matrix
    cluster_transition = np.zeros((n_clusters, n_clusters))
    
    # Convert transition matrix to dense if sparse
    tm = tm.A if issparse(tm) else tm
    
    # Compute transitions between clusters
    for i, source_cluster in enumerate(unique_clusters):
        source_indices = cluster_to_indices[source_cluster]
        n_source = len(source_indices)
        for j, target_cluster in enumerate(unique_clusters):
            target_indices = cluster_to_indices[target_cluster]
            total_transitions = tm[source_indices][:, target_indices].sum()
            cluster_transition[i, j] = total_transitions / n_source
    
    # Normalize rows to sum to 1
    cluster_transition = cluster_transition / cluster_transition.sum(axis=1, keepdims=True)
    
    # Convert to DataFrame with cluster labels
    cluster_transition_df = pd.DataFrame(
        cluster_transition,
        index=unique_clusters,  # Source clusters (rows)
        columns=unique_clusters  # Target clusters (columns)
    )
    
    return cluster_transition_df
```

```python
import pandas as pd
import numpy as np
from scipy.sparse import issparse

def add_avg_transition_and_sort(adata, tm):
    """
    Adds a new column to adata.obs with the average transition probability (excluding self-transitions)
    and sorts adata by this column in descending order.

    Parameters:
        adata: AnnData object.
        tm: Transition matrix (n_cells x n_cells).

    Returns:
        adata: Sorted AnnData object with a new column 'avg_transition' in adata.obs.
    """
    # Ensure tm is a dense array
    tm = tm.A if issparse(tm) else tm

    # Number of cells
    n_cells = tm.shape[0]

    # Initialize a list to store the average transition probabilities
    avg_transitions = []

    # Iterate over each cell
    for i in range(n_cells):
        # Get transition probabilities for cell i, excluding self-transition
        cell_transitions = np.delete(tm[i, :], i)  # Exclude the diagonal (self-transition)
        
        # Compute the average transition probability
        avg_transitions.append(np.mean(cell_transitions))

    # Add the average transition probabilities to adata.obs
    adata.obs['avg_transition'] = avg_transitions

    # Sort adata by the new column in descending order
    adata = adata[adata.obs['avg_transition'].sort_values(ascending=False).index]

    return adata
```

```python
import scanpy as sc
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def align_and_plot(adata1, adata2, obs_key, palette=None):
    # Ensure the obs columns are of type 'category'
    adata1.obs[obs_key] = adata1.obs[obs_key].astype('category')
    adata2.obs[obs_key] = adata2.obs[obs_key].astype('category')

    # Find unique categories from both AnnData objects
    categories = sorted(set(adata1.obs[obs_key].unique()).union(set(adata2.obs[obs_key].unique())))
    
    # Generate a color palette if not provided
    if not palette:
        cmap = plt.get_cmap('tab20')
        colors = cmap.colors
        palette = {cat: colors[i % len(colors)] for i, cat in enumerate(categories)}
    
    # Apply the color palette to the AnnData objects
    adata1.uns[obs_key + '_colors'] = [palette[cat] for cat in adata1.obs[obs_key].cat.categories]
    adata2.uns[obs_key + '_colors'] = [palette[cat] for cat in adata2.obs[obs_key].cat.categories]

    return adata1, adata2
```

```python
def get_predictions(
    current_time_point,
    future_time_point,
    subset=None,
    use_recs=False,
    ):
    
    rename_dict_inv = {v:k for k,v in rename_dict.items()}
    
    hparams['growing_neurons']['prior_mappers']['mappers']['time_mapper']['collapse_name'] = rename_dict_inv[future_time_point]
    outs = trainer.get_outputs(dataloader='all')
    
    ctypes = outs['predType']
    ctypes_str = [trainer.dataset.label_mapping['predType'][x] for x in ctypes]
    ctypes_str = np.array(ctypes_str)
    
    time = outs['time']
    time_str = [trainer.dataset.label_mapping['time'][x] for x in time]
    time_str = np.array(time_str)
    
    preds = outs['reconstructed_collapse']

    if use_recs:
        recs = outs['reconstructed']
    else:
        hparams['growing_neurons']['prior_mappers']['mappers']['time_mapper']['collapse_name'] = rename_dict_inv[current_time_point]
        outs = trainer.get_outputs(dataloader='all')
        recs = outs['reconstructed']

    obs = {
        'Time': time_str,
        'Cell type': ctypes_str
    }

    layers = {
        'spliced': recs,
        'unspliced': preds
    }
    
    adata_preds = sc.AnnData(
        X=recs,
        obs=obs,
        layers=layers
    )
    new_order = list(rename_dict.values())
    
    adata_preds.obs['Time'] = adata_preds.obs['Time'].astype('category')
    adata_preds.obs['Time'] = adata_preds.obs['Time'].cat.rename_categories(rename_dict)
    adata_preds.obs['Time'] = adata_preds.obs['Time'].cat.reorder_categories(new_order, ordered=True)

    if use_recs:
        adata_preds = adata_preds[adata_preds.obs['Time'] == current_time_point]
    else:
        adata_preds = adata_preds[adata_preds.obs['Time'] == 'Day 0']

    if subset:
        adata_preds = adata_preds[adata_preds.obs['Cell type'].isin(subset)]

    adata_preds.var_names = trainer.dataset.data.var_names.copy()
    return adata_preds
```

```python
"""
Compute cell-level distances between current (spliced) and future (unspliced) states 
in PCA space, identify top-affected cells, and visualize results.

Assumptions:
- `adata.layers['spliced']` and `adata.layers['unspliced']` hold matching gene-expression matrices
  (current vs. counterfactual future).
- `adata.obs['cell_type']` annotates each cell’s type.

This pipeline:
1. Projects both layers into a shared PCA embedding.
2. Computes L2 distances for each cell between its current and future representation.
3. Flags the top-X% most-affected cells per distance magnitude.
4. Plots the distance distribution and per–cell-type proportions of affected cells.
"""

import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt


def compute_l2_distances(adata, n_pcs=20):
    """
    Compute L2 distance between `spliced` and `unspliced` layers in PCA space.

    Adds `obs['l2_distance']` to `adata`.

    Parameters
    ----------
    adata : AnnData
        Must contain matching `layers['spliced']` and `layers['unspliced']`.
    n_pcs : int, default 20
        Number of principal components to compute.

    Returns
    -------
    AnnData
        The same AnnData with `obs['l2_distance']` populated.
    """
    # Stack current/future for joint PCA
    X = np.vstack([adata.layers['spliced'], adata.layers['unspliced']])
    temp = sc.AnnData(X, var=adata.var.copy())
    sc.pp.pca(temp, n_comps=n_pcs)

    # Split back
    n = adata.n_obs
    X_pca = temp.obsm['X_pca']
    current = X_pca[:n]
    future  = X_pca[n:]
    #future = adata.layers['unspliced']
    #current = adata.layers['spliced']
    # L2 distance per cell
    adata.obs['l2_distance'] = np.linalg.norm(future - current, axis=1)
    return adata


def identify_top_affected(adata, top_percentile=95):
    """
    Flag the top-percentile cells by `l2_distance` as 'most_affected'.

    Adds `obs['most_affected']` boolean column.

    Parameters
    ----------
    adata : AnnData
        Must have `obs['l2_distance']`.
    top_percentile : float in (0,100), default 95
        Percentile threshold for flagging.

    Returns
    -------
    AnnData
        Same AnnData with `obs['most_affected']` set.
    """
    thr = np.percentile(adata.obs['l2_distance'], top_percentile)
    adata.obs['most_affected'] = adata.obs['l2_distance'] >= thr
    adata.obs['most_affected'] = ['True' if x == True else 'False' for x in adata.obs['most_affected'].tolist()]
    adata.obs['most_affected'] = adata.obs['most_affected'].astype('category')
    return adata


def plot_celltype_proportions(adata, cell_type_key='cell_type'):
    """
    Plot stacked bar of the proportion of 'most_affected' cells per cell type.
    """
    prop = adata.obs.groupby(cell_type_key)['most_affected']    \
                   .mean()  # fraction affected per type

    plt.figure(figsize=(8,4.5))
    prop.plot(kind='bar', edgecolor='black', width=0.7)
    plt.xlabel('Cell type')
    plt.ylabel('Proportion of most-affected cells')
    plt.title('Most-affected cells by cell type')
    plt.tight_layout()
    plt.show()
```

```python
import requests
import json

def get_enrichr_results(gene_list, database):
    ENRICHR_URL = 'http://maayanlab.cloud/Enrichr/addList'
    genes_str = '\n'.join(gene_list)
    description = 'Example gene list'
    payload = {
        'list': (None, genes_str),
        'description': (None, description)
    }

    response = requests.post(ENRICHR_URL, files=payload)
    if not response.ok:
        raise Exception('Error analyzing gene list')

    data = json.loads(response.text)
    user_list_id = data['userListId']

    ENRICHR_URL = f'http://maayanlab.cloud/Enrichr/enrich?userListId={user_list_id}&backgroundType={database}'
    response = requests.get(ENRICHR_URL)
    if not response.ok:
        raise Exception('Error fetching enrichment results')

    data = json.loads(response.text)
    return pd.DataFrame(data[database], columns=['Rank', 'Term', 'P-value', 'Z-score', 'Combined Score', 'Overlapping Genes', 'Adjusted P-value', 'Old P-value', 'Old Adjusted P-value'])
```
