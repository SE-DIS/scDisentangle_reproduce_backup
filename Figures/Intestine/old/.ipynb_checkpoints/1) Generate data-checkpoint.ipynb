```python
seed_nb = 42
yaml_path = 'configs/disentangle_donor_id.yaml'
weights_path =  'weights/MIG_BINNED_dis_latent_stack_donor_id_train'
counterfactual_dict={}
n_latent_size = 16

# Change this line in loop:
    # covariates_batch = {'condition': torch.zeros(_adata.shape[0]).long().to(trainer.device)}
```

```python
set_seed(seed_nb)
```

```python
trainer = get_trainer(yaml_path, wandb_log=False)
```

```python
trainer.load_weights(weights_path)
```

```python
counterfactual_dict={'donor_id': 'A34 (417C)'}
```

```python
adata = trainer.predict(
    trainer.dataset.data.copy(), 
    counterfactual_dict=counterfactual_dict, 
    bs=256
)
```

```python
new_adata = None
disentangled_levels = []

for idx in tqdm(range(n_latent_size)):
    
    
    library = torch.log(torch.tensor(adata.X.sum(1)).unsqueeze(1)).to(trainer.device)
    
    dis_latent_stack_progressive = adata.obsm['dis_latent_stack'].copy()
    #dis_latent_stack_degressive = adata.obsm['dis_latent_stack'].copy()
    #dis_latent_stack_isolated = adata.obsm['dis_latent_stack'].copy()
    #dis_latent_stack = adata.obsm['dis_latent_stack'].copy()
    
    #cat_latent_stack_collapse = adata.obsm['cat_latent_stack_collapse'].copy()
    
    if idx != n_latent_size-1:
        dis_latent_stack_progressive[:, idx+1:] = dis_latent_stack_progressive[:, idx+1:].mean(axis=0)#.repeat(dis_latent_stack.shape[0], 0)
        #dis_latent_stack_degressive[:, :n_latent_size-1-idx+1] = dis_latent_stack_degressive[:, :n_latent_size-1-idx+1].mean(axis=0)
        #disentangled_levels.append(dis_latent_stack[:, :idx+1])

        #if idx != 0:
        #    dis_latent_stack_isolated[:, :idx] = dis_latent_stack_isolated[:, :idx].mean(axis=0)
        #    dis_latent_stack_isolated[:, idx+1:] = dis_latent_stack_isolated[:, idx+1:].mean(axis=0)

        #else:
        #    dis_latent_stack_isolated[:, idx+1:] = dis_latent_stack_isolated[:, idx+1:].mean(axis=0)
    else:
        pass
        #disentangled_levels.append(dis_latent_stack)
        #dis_latent_stack_isolated[:, :idx] = dis_latent_stack_isolated[:, :idx].mean(axis=0)
        
    #cat_latent_stack_collapse[:, :diff] = dis_latent_stack.copy()

    indx = trainer.dataset.reverse_label_mapping['donor_id']['A34 (417C)']
    covariates_batch = {'donor_id': (torch.ones(adata.shape[0]) * indx).long().to(trainer.device)}

    dis_latent_prog = [ torch.tensor(dis_latent_stack_progressive[:, x]).unsqueeze(1).to(trainer.device) for x in range(dis_latent_stack_progressive.shape[1]) ]

    #dis_latent_isolated = [ torch.tensor(dis_latent_stack_isolated[:, x]).unsqueeze(1).to(trainer.device) for x in range(dis_latent_stack_isolated.shape[1]) ]
    
    counterfactual_latent = trainer.get_counterfactuals(
                        x_inp=torch.tensor(dis_latent_stack_progressive).to(trainer.device),
                        variables=covariates_batch,
                        dis_latent=dis_latent_prog,
                        counterfactual_dict={},
                        suffixe='',
                        )

    counterfactual_recs = trainer.get_recs(
                        decoder_name='decoder',
                        decoder_input= counterfactual_latent['map_latent_summed'],
                        px_name='px_r',
                        library=library,
                        suffixe='_collapse',
                    )

    map_latent_summed = counterfactual_latent['map_latent_summed'].cpu().detach().numpy()
    reconstructed_collapse = counterfactual_recs['reconstructed_collapse'].cpu().detach().numpy()

    # _adata.obsm['dis_latent_individual'] = adata.obsm['dis_latent_stack'][:, idx].copy().reshape(-1, 1)
    # _adata.obsm['dis_latent_stack'] = adata.obsm['dis_latent_stack']
    # _adata.obsm['dis_latent_progressive'] = dis_latent_stack_progressive
    # _adata.obsm['dis_latent_stack_degressive'] = dis_latent_stack_degressive
    # _adata.obsm['map_latent_summed'] = map_latent_summed
    # _adata.obsm['cat_latent_stack_collapse'] = cat_latent_stack_collapse
    # _adata.obsm['reconstructed_collapse'] = reconstructed_collapse
    # _adata.obsm['disentangled_levels'] = np.concatenate(disentangled_levels, axis=1)

    _adata = sc.AnnData(
        X=reconstructed_collapse, 
        obsm = {'map_latent_summed': map_latent_summed},
        obs=adata.obs.copy()
    )
    _adata.obs['level'] = idx

    # counterfactual_latent_isolated = trainer.get_counterfactuals(
    #                     x_inp=torch.tensor(dis_latent_stack_isolated).to(trainer.device),
    #                     variables=covariates_batch,
    #                     dis_latent=dis_latent_isolated,
    #                     counterfactual_dict={'Batch_Library': '5.0'},
    #                     suffixe='',
    #                     )

    # counterfactual_recs_isolated = trainer.get_recs(
    #                     decoder_name='decoder',
    #                     decoder_input= counterfactual_latent_isolated['map_latent_summed'],
    #                     px_name='px_r',
    #                     library=library,
    #                     suffixe='_collapse',
    #                 )

    # map_latent_summed_isolated = counterfactual_latent_isolated['map_latent_summed'].cpu().detach().numpy()
    # reconstructed_collapse_isolated = counterfactual_recs_isolated['reconstructed_collapse'].cpu().detach().numpy()

    # _adata.obsm['dis_latent_stack_isolated'] = dis_latent_stack_isolated
    # _adata.obsm['map_latent_summed_isolated'] = map_latent_summed_isolated
    # _adata.obsm['reconstructed_collapse_isolated'] = reconstructed_collapse_isolated
    
    if new_adata is None:
        new_adata = _adata.copy()
    else:
        new_adata = ad.concat([new_adata, _adata])
```
