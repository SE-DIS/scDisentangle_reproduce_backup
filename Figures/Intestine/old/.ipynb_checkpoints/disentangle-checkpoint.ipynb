```python
import warnings
warnings.filterwarnings('ignore')
```

# Imports

```python
import torch
import scanpy as sc
import numpy as np
import pandas as pd
from scdisentangle.train.tools import get_trainer, set_seed
```

# Set seed

```python
seed_nb = 42

set_seed(seed_nb)
```

```python
yaml_path = 'configs/disentangle_donor_id.yaml'
```

# Get trainer

```python
trainer = get_trainer(yaml_path, wandb_log=True)
```

# Train

```python
trainer.train()
```

# Load best weights

```python
weights_path = 'weights/MIG_BINNED_dis_latent_stack_donor_id_train'
trainer.load_weights(weights_path)
```

# Get predictions (batch corrected)

```python
counterfactual_dict={'donor_id': 'A34 (417C)'}

adata = trainer.predict(
    trainer.dataset.data.copy(), 
    counterfactual_dict=counterfactual_dict, 
    bs=256
)
```

# Plot batch corrected latent

```python
map_latent_collapse = sc.AnnData(
    X = adata.obsm['dis_latent_stack'], 
    obs = adata.obs.copy()
)
```

```python
sc.pp.neighbors(map_latent_collapse)
sc.tl.umap(map_latent_collapse)
```

```python
sc.pl.umap(map_latent_collapse, color=['author_cell_type', 'donor_id'])
```
